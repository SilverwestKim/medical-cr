```python
k=driver.find_elements_by_class_name("btnText")
l=len(k)
```

```python
text=[]
```

```python
for i in range(l):
    k[i].click()
    driver.switch_to.window(driver.window_handles[-1])
    time.sleep(2)
    sub=driver.find_element_by_css_selector('xmp')
    a=sub.text
    text.append(a)
    driver.close()
    driver.switch_to.window(driver.window_handles[0])
    time.sleep(2)
```

```python
data=pd.DataFrame({'text':text})
```
